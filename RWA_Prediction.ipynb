```python
#ENGINEERING DES FEATURES SPÉCIFIQUES AU CONTEXTE BANCAIRE

print("\nEngineering des features")

# Ratios financiers importants pour RWA
if 'outstanding_amount_in_local_currency' in df_reduit.columns and 'total_customer_outstanding_amount' in df_reduit.columns:
    df_reduit['exposure_concentration_ratio'] = (
        df_reduit['outstanding_amount_in_local_currency'] / 
        (df_reduit['total_customer_outstanding_amount'] + 1e-6)
    )

if 'total_customer_provision_base' in df_reduit.columns and 'total_customer_outstanding_amount' in df_reduit.columns:
    df_reduit['provision_coverage_ratio'] = (
        df_reduit['total_customer_provision_base'] / 
        (df_reduit['total_customer_outstanding_amount'] + 1e-6)
    )

if 'net_outstanding_post_cover' in df_reduit.columns and 'outstanding_amount_in_local_currency' in df_reduit.columns:
    df_reduit['coverage_effectiveness'] = (
        df_reduit['net_outstanding_post_cover'] / 
        (df_reduit['outstanding_amount_in_local_currency'] + 1e-6)
    )

if 'weighting' in df_reduit.columns:
    df_reduit['high_risk_weighting'] = (df_reduit['weighting'] > df_reduit['weighting'].median()).astype(int)
    df_reduit['weighting_squared'] = df_reduit['weighting'] ** 2
    df_reduit['weighting_log'] = np.log1p(df_reduit['weighting'])

# Interactions entre variables clés
if 'is_provisionable' in df_reduit.columns and 'provision_rate' in df_reduit.columns:
    df_reduit['provision_risk_score'] = df_reduit['is_provisionable'] * df_reduit['provision_rate']

if 'is_guarantable' in df_reduit.columns and 'total_customer_guarantee_base' in df_reduit.columns:
    df_reduit['guarantee_effectiveness'] = df_reduit['is_guarantable'] * df_reduit['total_customer_guarantee_base']

# Binning des montants
for col in ['outstanding_amount_in_local_currency', 'total_customer_outstanding_amount']:
    if col in df_reduit.columns:
        df_reduit[f'{col}_quartile'] = pd.qcut(df_reduit[col], q=4, labels=['Low', 'Medium', 'High', 'Very_High'], duplicates='drop')

print(f"Features ajoutées. Nouvelles dimensions: {df_reduit.shape}")
```

```python
from sklearn.linear_model import RidgeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.ensemble import ExtraTreesClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# --- KNN Classifier ---
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
y_pred_knn = knn.predict(x_test)

print("📌 KNN Classifier Results:")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
print("Classification Report:\n", classification_report(y_test, y_pred_knn))
print("Accuracy Score:", accuracy_score(y_test, y_pred_knn))

# --- Ridge Classifier ---
ridge = RidgeClassifier(alpha=1.0)
ridge.fit(x_train, y_train)
y_pred_ridge = ridge.predict(x_test)

print("\n📌 Ridge Classifier Results:")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_ridge))
print("Classification Report:\n", classification_report(y_test, y_pred_ridge))
print("Accuracy Score:", accuracy_score(y_test, y_pred_ridge))




# --- Extra Trees Classifier ---
et = ExtraTreesClassifier(n_estimators=100, random_state=42)
et.fit(x_train, y_train)
y_pred_et = et.predict(x_test)
print("\n📌 Extra Trees Classifier Results:")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_et))
print("Classification Report:\n", classification_report(y_test, y_pred_et))
print("Accuracy Score:", accuracy_score(y_test, y_pred_et))



```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_conf_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title(title)
    plt.show()

plot_conf_matrix(y_test, y_pred_knn, "KNN Confusion Matrix")
plot_conf_matrix(y_test, y_pred_et, "Extra Trees Confusion Matrix")
plot_conf_matrix(y_test, y_pred_ridge, "Ridge Confusion Matrix")
```

```python
import matplotlib.pyplot as plt
from sklearn.calibration import calibration_curve
from sklearn.metrics import brier_score_loss
import numpy as np

# Use SVC with probability=True
models = {
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "ExtraTrees": ExtraTreesClassifier(n_estimators=100, random_state=42),
    "ridge" : RidgeClassifier(alpha=1.0),

    
}
plt.figure(figsize=(10, 7))

for name, model in models.items():
    model.fit(x_train, y_train)
    if hasattr(model, "predict_proba"):
        prob_pos = model.predict_proba(x_test)[:, 1]
    elif hasattr(model, "decision_function"):
        # Use sigmoid to convert decision_function output to probability
        decision = model.decision_function(x_test)
        prob_pos = 1 / (1 + np.exp(-decision))
    else:
        raise AttributeError(f"{name} does not support probability prediction.")

    # Courbe de calibrage
    fraction_of_positives, mean_predicted_value = calibration_curve(
        y_test, prob_pos, n_bins=10, strategy='uniform'
    )

    plt.plot(mean_predicted_value, fraction_of_positives, "o-", label=name)
    print(f"{name} Brier score: {brier_score_loss(y_test, prob_pos):.4f}")

# ➕ Ligne diagonale = modèle parfaitement calibré
plt.plot([0, 1], [0, 1], "k:", label="Perfectly calibrated")

plt.xlabel("Probabilité prédite")
plt.ylabel("Probabilité réelle (observée)")
plt.title("Courbe de calibrage des modèles")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import RidgeClassifier
import numpy as np

# Define models
models = {
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "ExtraTrees": ExtraTreesClassifier(n_estimators=100, random_state=42),
    "ridge" : RidgeClassifier(alpha=1.0)

}

plt.figure(figsize=(10, 7))

for name, model in models.items():
    model.fit(x_train, y_train)
    
    if hasattr(model, "predict_proba"):
        probas = model.predict_proba(x_test)[:, 1]
    elif hasattr(model, "decision_function"):
        decision = model.decision_function(x_test)
        probas = 1 / (1 + np.exp(-decision))
    else:
        raise AttributeError(f"{name} does not support probability prediction.")
    
    fpr, tpr, _ = roc_curve(y_test, probas)
    roc_auc = auc(fpr, tpr)
    
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

# ➕ Ligne diagonale = modèle aléatoire
plt.plot([0, 1], [0, 1], 'k--', label='Random')

plt.xlabel("Taux de faux positifs (FPR)")
plt.ylabel("Taux de vrais positifs (TPR)")
plt.title("Courbe ROC des modèles")
plt.legend(loc="lower right")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#KAPPA DE CHAQUE MODÈLE
from sklearn.metrics import cohen_kappa_score
# Calculer le Kappa pour chaque modèle
models = {
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "ExtraTrees": ExtraTreesClassifier(n_estimators=100, random_state=42),
    "ridge" : RidgeClassifier(alpha=1.0)

}
kappa_scores = {}
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    kappa_scores[name] = cohen_kappa_score(y_test, y_pred)
# Afficher les scores Kappa
for name, kappa in kappa_scores.items():
    print(f"{name} Kappa Score: {kappa:.4f}")     

#MCC DE CHAQUE MODÈLE
from sklearn.metrics import matthews_corrcoef
# Calculer le MCC pour chaque modèle
mcc_scores = {}
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    mcc_scores[name] = matthews_corrcoef(y_test, y_pred)
# Afficher les scores MCC
for name, mcc in mcc_scores.items():
    print(f"{name} MCC Score: {mcc:.4f}")   
```
